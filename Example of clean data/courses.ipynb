```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import re
p="https://cs.nyu.edu/dynamic/courses/schedule/"
p2=requests.get(p)
course = BeautifulSoup(p2.text,"html.parser")
m="https://cs.nyu.edu/dynamic/courses/catalog/"
m2=requests.get(m)
category = BeautifulSoup(m2.text,"html.parser")
```

```python
table_course = list()

for li in course.find_all('li'):
    trow={}
    i = 0   
    for span in li.find_all('span'):
        i+=1
        
        if i<5:
            
            # a_tag = span.find('a')
            if i == 1:
                a_tag = span.find('a')
                text_before_br = span.contents[0].strip()
                if a_tag:
                    Id = a_tag.text.strip()
                    trow['Number-Section']=Id
                    
                elif text_before_br: 
                    Id = text_before_br
                    trow['Number-Section']=Id
                    
                else:    
                    ID = span.text.strip()
                    trow['Number-Section'] = ID
                    
                    
            if i == 2:
                name = span.text.strip()
                trow['Name'] = name
                
            if i == 3:
                instr = span.text.strip()
                trow['Instructor'] = instr
                
            if i == 4:
                time = span.text.strip()
                trow['Time'] = time
                
               
            if i == 1:
                match = re.search(r'CSCI-(GA|UA)\.\d{4}', span.text)
                if match:
                    trow['Number'] = match.group()
                    table_course.append(trow)
                    
                table_course.append(trow)

            
            
            

course_schedule =pd.DataFrame(table_course)
columns = [col for col in course_schedule.columns if col != 'Number'] + ['Number']
course_schedule = course_schedule[columns]
course_schedule
```

```python
table_category = list()
for li in category.find_all('li'):
    throw= {}
    i = 0
    for p in li.find_all('p'):
        i += 1
        if i == 1:
            match = re.search(r'(CSCI|FRSEM)-(GA|UA)\.\d{4}', p.text)
            if match:
                throw['Number'] = match.group()
                
        if i == 2:
            match = re.search(r'\b(\d+)\b', p.text)
            if match:
                throw['Points'] = match.group(1)
                
        if i == 3:
            match = re.search(r'Prerequisites:\s*(.*)', p.text)
            if match:
                throw['Prereqs'] = match.group(1)
                table_category.append(throw)
            
            
       
        

        
course_category =pd.DataFrame(table_category)
columns = [col for col in course_category.columns if col != 'Points'] + ['Points']
course_category = course_category[columns]
course_category
```

```python

merged_df = pd.merge(course_schedule, course_category, on='Number', how='left')
final_df = merged_df[['Number', 'Name', 'Instructor', 'Time', 'Prereqs', 'Points']]

final_df
```
